# Web Scraping (html)

```python
# Import Necessary Libraries
import requests
from bs4 import BeautifulSoup
```

## Scraping Books Website

```python
# Check website response
# Note this url is for mystery page
response = requests.get('http://books.toscrape.com/catalogue/category/books/mystery_3/index.html')
response
```

```python
# Get website data
s = BeautifulSoup(response.text, 'html.parser')
s
```

```python
# To get specific line in page we can get it by providing Tag name and class name as the following
s.find_all('div', 'image_container')
```

```python
# Book Name first value
# To get tag inside a tag we use '.' and to get any thing inside a tag like (alt = "Sharp Objects") we use ['alt'] to get its value "Sharp Objects"
s.find_all('article', 'product_pod')[0].div.a.img['alt']
```

```python
# Book Name loop
# We always loop on a list first, then scrap inside the for loop
for book in s.find_all('article', 'product_pod'):
    
    print(book.div.a.img['alt'])
```

```python
# Rating first value
s.find_all('article', 'product_pod')[0].p['class'][1]
```

```python
# Rating loop
for rate in s.find_all('article', 'product_pod'):
    
    print(rate.p['class'][1])
```

```python
# Price first value
s.find_all('article', 'product_pod')[0].find_all('div')[1].p.text
```

```python
# Price loop
for price in s.find_all('article', 'product_pod'):
    
    print(price.find_all('div')[1].p.text)
```

```python
# Availability first value
s.find_all('article', 'product_pod')[0].find_all('div')[1].find_all('p')[1].text.strip()
```

```python
# Availability loop
for avail in  s.find_all('article', 'product_pod'):
    
    print(avail.find_all('div')[1].find_all('p')[1].text.strip())
```

```python
# Create table to fill data in using pandas library
import pandas as pd

df = pd.DataFrame(columns= ['Book Name', 'Rating', 'Price', 'Availability'])

df
```

```python
# We create one for loop that includes all we need to scrap
for idx, article in enumerate(s.find_all('article', 'product_pod')):
    
    book_name = article.div.a.img['alt']
    
    rating = article.p['class'][1]
    
    price = article.find_all('div')[1].p.text
    
    avail = article.find_all('div')[1].find_all('p')[1].text.strip()
    
    df.loc[idx, 'Book Name'] = book_name
    
    df.loc[idx, 'Rating'] = rating
    
    df.loc[idx, 'Price'] = price
    
    df.loc[idx, 'Availability'] = avail
    
df
```

```python
# Then we can save file as csv
df.to_csv('Books_scrap1.csv')
```

```python
# Scrapping all pages
# This url is for home page
response = requests.get('http://books.toscrape.com/index.html')
response
```

```python
s = BeautifulSoup(response.text, 'html.parser')
```

```python
# To get any page complete link, we need to add it to this part
'http://books.toscrape.com/' +  s.find('ul', 'nav nav-list').li.ul.a['href']
```

```python
for page in s.find('ul', 'nav nav-list').ul.find_all('li'):
    
    print('http://books.toscrape.com/' + page.a['href'])
```

```python
# Scrap all categories pages

i1 = 0
i2 = 0
i3 = 0
i4 = 0

for page in s.find('ul', 'nav nav-list').ul.find_all('li'):
    
    page_url = 'http://books.toscrape.com/' + page.a['href']
    

    response = requests.get(page_url)
    
    s = BeautifulSoup(response.text, 'html.parser')
    
    
    for article in s.find_all('article', 'product_pod'):
    
        book_name = article.div.a.img['alt']

        rating = article.p['class'][1]

        price = article.find_all('div')[1].p.text

        avail = article.find_all('div')[1].find_all('p')[1].text.strip()

        df.loc[i1, 'Book Name'] = book_name

        i1 += 1

        df.loc[i2, 'Rating'] = rating

        i2 += 1

        df.loc[i3, 'Price'] = price

        i3 += 1

        df.loc[i4, 'Availability'] = avail

        i4 += 1
    
df
```
